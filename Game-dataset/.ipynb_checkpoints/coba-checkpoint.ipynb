# Perbandingan Rating 

```python
bagus = gaming[gaming['average rating'] >=4]
kurang = gaming[gaming['average rating'] <= 3]
```

```python
import matplotlib.pyplot as plt 
warna = ['cyan', 'yellow']
deskripsi = ["Baik", "Kurang"]
produk = [bagus['title'].count(),kurang['title'].count()]
expose = [0.2, 0]
plt.title('Perbandingan rating produk')
plt.pie(produk,
        labels=deskripsi,
        autopct= '%1.1f%%',
        colors = warna,
        shadow = True,
        explode = expose) 

plt.show()

baik = bagus['title'].count()
belum = kurang['title'].count()
print('Jumlah game dengan rating >= 4 :', baik)
print('Jumlah game dengan rating <= 3 :', belum)
```
